```python
from openai import AzureOpenAI  # Importe la classe principale pour interagir avec les modèles OpenAI hébergés sur le cloud Azure.
from azure.core.credentials import AzureKeyCredential  # Importe l'outil de gestion sécurisée des clés d'authentification pour les services Azure.


endpoint = "https://my-embidding-pascal.openai.azure.com/"  # Définit l'adresse URL unique (le point d'entrée) de ton instance de service Azure OpenAI.
model_name = "text-embedding-3-small"  # Définit le nom technique du modèle (ici, un modèle qui transforme du texte en vecteurs numériques).
deployment = "my-beddinglab-pascal"  # Définit le nom personnalisé que tu as donné à ce modèle lors de son déploiement dans le portail Azure.


api_version = "2024-02-01"  # Définit la version spécifique de l'API à utiliser pour garantir la compatibilité des fonctionnalités.


client = AzureOpenAI(  # Initialise l'objet 'client' qui servira de passerelle pour envoyer tes requêtes vers Azure.)
    api_version="2024-12-01-preview",  # Spécifie la version de l'API (note : elle écrase la variable définie plus haut).
    azure_endpoint=endpoint,  # Transmet l'URL de ton service pour que le code sache vers quel serveur envoyer les données.
    api_key= api_key # Transmet ta clé secrète pour prouver que tu as l'autorisation d'utiliser ce service payant.)
)
```

```python
import numpy as np  # Importe la bibliothèque de calcul numérique pour manipuler les listes de nombres (vecteurs) efficacement.
import matplotlib.pyplot as plt  # Importe l'outil de création de graphiques (même si non utilisé ici, utile pour visualiser les données).
import seaborn as sns  # Importe une surcouche à matplotlib pour rendre les futurs graphiques plus esthétiques.
from scipy.spatial.distance import cosine  # Importe la fonction spécifique pour calculer l'écart "angulaire" entre deux vecteurs.
```

```python
# --- LES DEUX PHRASES ---
phrase1 = "Le chat dort sur le tapis."  # Définit la première chaîne de caractères à analyser.
phrase2 = "Un félin se repose sur la moquette."  # Définit la seconde phrase, sémantiquement proche mais avec des mots différents.


def get_embedding(text):  # Crée une fonction personnalisée pour simplifier l'appel à l'IA d'Azure.
    # Envoie le texte à Azure, récupère la réponse du modèle de déploiement, et extrait uniquement la liste de nombres (l'embedding).
    return client.embeddings.create(input=text, model=deployment).data[0].embedding
```

```python
# 1. Génération des embeddings
emb1 = get_embedding(phrase1)  # Transforme la phrase 1 en une liste de 1536 nombres (coordonnées mathématiques).
emb2 = get_embedding(phrase2)  # Transforme la phrase 2 en une liste de nombres identique en taille pour pouvoir les comparer.
print(f"Embeddings de la phrase 1 : {emb1}")  # Affiche la liste de nombres pour vérifier que la conversion a réussi.
print(f"Embeddings de la phrase 2 : {emb2}")  # Affiche la seconde liste de nombres.
```

```python
# 2. Calcul de la similarité (Objectif du labo)
similarity = 1 - cosine(emb1, emb2)  # Calcule la "similarité cosinus" : 1 signifie que les sens sont identiques, 0 qu'ils sont opposés.
# Pourquoi '1 - cosine' : La fonction 'cosine' mesure la distance (l'écart) ; on la soustrait à 1 pour obtenir la ressemblance.
print(f"Similarité sémantique : {similarity:.4f}")  # Affiche le score final arrondi à 4 décimales (ex: 0.9200).
```
